Exploratory Data Analysis

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
abc= sns.countplot(x='Gender',data=df)
plt.title("Gender Distribution")
for bars in abc.containers:
    abc.bar_label(bars)
```

```python
ax = sns.countplot(data = df, x = 'Age Group', hue = 'Gender')
plt.title('Age and gender distribution')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sns.countplot(x='Gender', data=df)
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
sns.histplot(x='Age',data=df)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
sns.countplot(x='Product_Category',data=df)
plt.title("Product_Category")
plt.xlabel('product')
plt.ylabel("count")
plt.xticks(rotation=90)# to rotate the labels of xaxes
```

```python
df['total_Sales']=df['Amount']*df['Orders']
total_sales_by_gender=df.groupby('Gender')['total_Sales'].sum().reset_index()
sns.barplot(x='Gender' , y="total_Sales",data=df)
plt.title("total_sales_by gender")
```

```python
sns.barplot(x="Product_Category", y="total_Sales", data=df)
plt.title("Total Sales by Product Category")
plt.xticks(rotation=90)
```

```python

total_sales = df.groupby('State')['total_Sales'].sum()

plt.figure(figsize=(8, 8))
plt.pie(total_sales, labels=total_sales.index, autopct='%1.1f%%', startangle=140)
plt.title('Total Orders Over States')
plt.show()
```

```python
sns.boxplot(data=df, x='Orders',y='State')
plt.title('BOX Plot')
plt.xlabel('Orders')
plt.ylabel('state')
plt.xticks(rotation=45)  
plt.show()
```
